## 04. Transposing data

```python
df_dep = pd.read_csv(r'C:\Users\efens\cf_tasks\2023-07 Instacard Basket Analysis\02 Data\021 Original data\departments.csv', index_col = False)
```

```python
df_dep.head(5)
```

```python
# transpose

df_dep.T
```

```python
# overwriting the df to a new transposed one

df_dep_t = df_dep.T
```

```python
df_dep_t
```

```python
# adding an index to a df

df_dep_t.reset_index()
```

```python
# using the 1st row for header
# iloc = “locate an integer”

new_header = df_dep_t.iloc[0]
```

```python
new_header
```

```python
# creating a new df that contains only rows beyound the first row

df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# adding a new header

df_dep_t_new.columns = new_header
df_dep_t_new
```
